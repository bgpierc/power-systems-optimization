```julia
using PyCall
using Dates
```

```julia
@pyimport pvlib
import pandas
import matplotlib.pyplot
```

```julia
latitude = 30.0  # Replace with your latitude
longitude = -90.0  # Replace with your longitude
tz = "Etc/GMT+6"  # Specify the timezone
location = pvlib.location.Location(latitude, longitude, tz=tz)
```

```julia
date = DateTime("2023-04-10T00:00:00")  # Specify the date and time
times = date:Minute(30):date + Dates.Day(1)
```

```julia
solar_position = pvlib.solarposition.get_solarposition(times, latitude, longitude)
```

```julia
tracker_angle = pvlib.tracking.singleaxis(solar_position[:apparent_zenith], solar_position[:azimuth],
    axis_tilt=0, axis_azimuth=180)
```

```julia
using PyPlot
plt.figure()
plt.plot(times, tracker_angle.tracker_theta, label="Tracker Angle")
plt.xlabel("Time")
plt.ylabel("Tracker Angle (degrees)")
plt.title("Single-Axis Solar Tracker Position")
plt.grid()
plt.legend()
plt.show()
```
